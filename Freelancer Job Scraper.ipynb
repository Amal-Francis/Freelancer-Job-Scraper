```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
```

```python
URL = 'https://www.freelancer.in/jobs'
r = requests.get(URL)
```

```python
r.content
```

```python
soup = BeautifulSoup(r.content, 'html.parser')
```

```python
print(soup.prettify())
```

```python
results = soup.find_all(class_='JobSearchCard-item')
```

```python
job={'Job':[], 'Time Remaining': [], 'Description':[], 'Requirement':[], 'Bid Price':[]}
```

```python
for item in results:
    job['Job'].append(item.find('a', class_='JobSearchCard-primary-heading-link').text.strip())
    job['Time Remaining'].append(item.find('span', class_='JobSearchCard-primary-heading-days').text.strip())
    job['Description'].append(item.find('p', class_='JobSearchCard-primary-description').text.strip().replace('\n', ''))
    job['Requirement'].append(item.find('div', class_='JobSearchCard-primary-tags').text.strip().replace('\n', ', '))
    try:
        job['Bid Price'].append(item.find('div', class_='JobSearchCard-primary-price').text.strip().replace('\n', ''))
    except AttributeError:
        job['Bid Price'].append(np.nan)
    
```

```python
Job_data=pd.DataFrame(job)
```
